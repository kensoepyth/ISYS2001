## Part 3: Building the Budget Tracker

Now for the real thing. Ask AI to help you write a program that:
- Takes the user's monthly income
- Takes up to 3 expenses
- Calculates what's left over
- Warns the user if the budget goes negative

> 💡 **Pro tip:** The more specific your prompt, the less clean-up you'll need to do afterwards. "Write a budget tracker" gets you something generic; describing the exact inputs, outputs, and edge cases gets you something closer to what you actually need.


````python
# Write your detailed prompt here:
"""
My Budget Tracker Prompt:
[Write your prompt here]
"""
````

Here is a detailed prompt you could use for an AI assistant to generate the budget tracker code:

```
Write a Python program that functions as a simple budget tracker. The program should:
1. Prompt the user to enter their monthly income as a numerical value.
2. Allow the user to enter up to three individual expenses. For each expense, it should ask for the expense name (string) and the expense amount (numerical value).
3. Calculate the total of all entered expenses.
4. Calculate the remaining budget by subtracting the total expenses from the monthly income.
5. Print a clear summary showing:
   - The monthly income.
   - Each expense entered (name and amount).
   - The total expenses.
   - The remaining budget.
6. If the remaining budget is negative, display a warning message indicating that the user has overspent and by how much.
7. Include basic input validation to ensure that income and expense amounts are positive numbers.
```

### Paste and Test Your AI-Generated Budget Tracker

````python
# AI-GENERATED BUDGET TRACKER CODE
# [Paste your code here]
````

````python
# AI-GENERATED BUDGET TRACKER CODE

def get_positive_float_input(prompt):
    while True:
        try:
            value = float(input(prompt))
            if value < 0:
                print("Value cannot be negative. Please enter a positive number.")
            else:
                return value
        except ValueError:
            print("Invalid input. Please enter a numerical value.")


print("\n--- Simple Budget Tracker ---\n")

# 1. Get user's monthly income
monthly_income = get_positive_float_input("Enter your monthly income: $")

# 2. Get up to 3 expenses
expenses = []
num_expenses_to_collect = 0
while True:
    try:
        num_expenses_to_collect = int(input("How many expenses (up to 3) would you like to enter? "))
        if 1 <= num_expenses_to_collect <= 3:
            break
        else:
            print("Please enter a number between 1 and 3.")
    except ValueError:
        print("Invalid input. Please enter a whole number.")

for i in range(num_expenses_to_collect):
    expense_name = input(f"Enter name for expense {i+1}: ")
    expense_amount = get_positive_float_input(f"Enter amount for {expense_name}: $")
    expenses.append({'name': expense_name, 'amount': expense_amount})

# 3. Calculate total expenses
total_expenses = sum(expense['amount'] for expense in expenses)

# 4. Calculate remaining budget
remaining_budget = monthly_income - total_expenses

# 5. Print summary
print("\n--- Budget Summary ---")
print(f"Monthly Income: ${monthly_income:,.2f}")
print("Expenses:")
for expense in expenses:
    print(f"  - {expense['name']}: ${expense['amount']:,.2f}")
print(f"Total Expenses: ${total_expenses:,.2f}")
print(f"Remaining Budget: ${remaining_budget:,.2f}")

# 6. Warn if budget is negative
if remaining_budget < 0:
    print(f"\n⚠️ Warning: You have overspent by ${abs(remaining_budget):,.2f} this month!")
````

````python
# ENHANCED AI-GENERATED BUDGET TRACKER CODE

def get_positive_float_input(prompt):
    while True:
        try:
            value = float(input(prompt))
            if value < 0:
                print("Value cannot be negative. Please enter a positive number.")
            else:
                return value
        except ValueError:
            print("Invalid input. Please enter a numerical value.")


print("\n--- Simple Budget Tracker ---\n")

# 1. Get user's monthly income
monthly_income = get_positive_float_input("Enter your monthly income: $")

# 2. Get up to 3 expenses
expenses = []
num_expenses_to_collect = 0
while True:
    try:
        num_expenses_to_collect = int(input("How many expenses (up to 3) would you like to enter? "))
        if 1 <= num_expenses_to_collect <= 3:
            break
        else:
            print("Please enter a number between 1 and 3.")
    except ValueError:
        print("Invalid input. Please enter a whole number.")

for i in range(num_expenses_to_collect):
    expense_name = input(f"Enter name for expense {i+1}: ")
    expense_amount = get_positive_float_input(f"Enter amount for {expense_name}: $")
    expenses.append({'name': expense_name, 'amount': expense_amount})

# 3. Calculate total expenses
total_expenses = sum(expense['amount'] for expense in expenses)

# 4. Calculate remaining budget
remaining_budget = monthly_income - total_expenses

# 5. Calculate percentage of income spent
percentage_spent = (total_expenses / monthly_income) * 100 if monthly_income > 0 else 0

# 6. Print summary
print("\n--- Budget Summary ---")
print(f"Monthly Income: ${monthly_income:,.2f}")
print("Expenses:")
for expense in expenses:
    print(f"  - {expense['name']}: ${expense['amount']:,.2f}")
print(f"Total Expenses: ${total_expenses:,.2f}")
print(f"Remaining Budget: ${remaining_budget:,.2f}")
print(f"Percentage of Income Spent: {percentage_spent:.2f}%")

# 7. Warn if budget is negative
if remaining_budget < 0:
    print(f"\n⚠️ Warning: You have overspent by ${abs(remaining_budget):,.2f} this month!")
````
